```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis, skew

def precompute_dataset_stats(df, name):
    rev_col = 'REVENUE'
    
    stats_dict = {
        'Dataset': name,
        'User_Count': len(df),
        'Conversion_Rate': (df[rev_col] > 0).mean(),
        'Mean_REVENUE': df[rev_col].mean(),
        'Max_REVENUE': df[rev_col].max(),
        'Skewness': skew(df[rev_col]),
        'Kurtosis': kurtosis(df[rev_col]),
    }
    
    # Revenue concentration 
    p99_threshold = df[rev_col].quantile(0.99)
    total_rev = df[rev_col].sum()
    top_1pct_rev = df[df[rev_col] >= p99_threshold][rev_col].sum()
    stats_dict['Top_1pct_Revenue_Share'] = top_1pct_rev / total_rev if total_rev > 0 else 0
    
    return stats_dict


# Precompute for each dataset
file_paths = ['Dataset1_outliers_ready.csv','Dataset2_outliers_ready.csv', 'Dataset3_outliers_ready.csv']
summary_reports = []
for i, path in enumerate(file_paths):
    try:
        df = pd.read_csv(path)
        name = path.replace('.csv', '')
        summary_reports.append(precompute_dataset_stats(df, name))
    except FileNotFoundError:
        print(f"File {path} not found. Skipping.")

# Output the outcome
precomputed_df = pd.DataFrame(summary_reports)
print("\n--- PRECOMPUTATION SUMMARY ---")
print(precomputed_df.to_string(index=False))
precomputed_df.to_csv('precomputation_results.csv', index=False)

```
